### **Dataframe with relevant columns**

```python
# To make it easier we will convert all the columns' name to lowercase

df.columns = [col.lower() for col in df]

#Now, we are going to get a new DataFrame removing all unimportant columns (own opinion)
relevant_df = df.drop(["actualelapsedtime","crselapsedtime","taxiin","taxiout"],axis=1)
relevant_df.head()

#In addition, we are going to convert the date in an individual column

relevant_df["date_of_flight"] = pd.to_datetime((relevant_df.year * 10000 + relevant_df.month * 100 + relevant_df.dayofmonth).apply(str),format = '%Y%m%d')

#change the date position and write it more lovely 

cols = list(relevant_df.columns) #Put the columns in a list
cols2 = cols[-1:] + cols[:-1] #Revert the columns like you prefer
new_df = relevant_df[cols2] #Create a new df with other word 
new_df

#Removing the columns that are not important for our data base

new_df = new_df.drop(["year","month","dayofmonth","dayofweek"], axis = 1)
new_df.head()
```

### **Create new columns (average flight speed, whether late or not ...)**

#### **Average flight speed**

```python
#We are going to calculate the average flight speed and add it into our data set

new_df["average_flight_speed(km/h)"] = ((new_df["distance"] * 1.60934) / (new_df["airtime"] / 60)).round(2)
new_df.head()
```

#### **Flight was late or not**

```python
#We are going to construct a dichotomous variable. 1 if is late 0 on the contrary
#We add the new variable in our data set

def delayedFligt(x):
    if x > 0:
        return 1
    else:
        return 0

new_df["delayed_flight"] = (new_df["arrdelay"].apply(delayedFligt))
new_df.head()
```

### **Table of airlines with the most accumulated arrears**

```python
#Before starting, I want to know the delays per airline

freq = pd.DataFrame(new_df.groupby(['uniquecarrier','delayed_flight']).size())
freq.head(6)
```

```python
#FIRST METHOD (using groupby)

airlineas_delays= pd.DataFrame(new_df.groupby('uniquecarrier')['delayed_flight'].agg(sum))
airlineas_delays=airlineas_delays.sort_values('delayed_flight',ascending=False) #Reordering
airlineas_delays.head()
```

```python
#SECOND METHOD (without groupby)

delays = new_df['delayed_flight'] == 1 #condition 
airlineas_delays2 = new_df[delays]
airlineas_delays2 = pd.DataFrame(airlineas_delays2['uniquecarrier'].value_counts())
airlineas_delays2.head()
```

### **What are the longest flights? And the most backward?**

#### **Longest Flights**

```python
#DISTANCE(MILES) #METHOD 1

first_step = new_df.groupby(['uniquecarrier','flightnum']).agg({'distance':max}) #We do a groupby and add the distance with the maximum values
second_step = first_step['distance'].groupby(level=0, group_keys=False) #We make a groupby for the variable distance
third_step = second_step.nlargest(2) #We only add the biggest number (two flights per company). Intention: see the other flights
x = pd.DataFrame(third_step.sort_values(ascending=False).head(10)) #Reordering the values and see the first five flights
x
```

```python
#TIME(MINUTES) #METHOD 2

new_df[['airtime', 'uniquecarrier','flightnum']].sort_values(by='airtime', ascending=False).nlargest(10, 'airtime')
#We do the same for the variable "airtime"
```

#### **Most backward**

```python
# First at all, I want to know the flight with most backward

new_df['arrdelay'].min()
```

```python
#DELAYS (METHOD 1)

first_step = new_df.groupby(['uniquecarrier','flightnum']).agg({'arrdelay':min}) #We do a groupby and add the "arrdelay" with the maximum values
second_step = first_step['arrdelay'].groupby(level=0, group_keys=False) #We make a groupbyfor the variable "arrdelay"
third_step = second_step.nsmallest(2) #We only add the smallest number 
x = pd.DataFrame(third_step.sort_values(ascending=True).head(10)) #Reordering the values and see the first five maximum delays
x
```

```python
#DELAYS (METHOD 2)

new_df[['arrdelay', 'uniquecarrier','flightnum']].sort_values(by='arrdelay', ascending=False).nsmallest(10, 'arrdelay')
#We do the same for the variable "arrdelay"
```

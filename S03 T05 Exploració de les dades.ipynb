### **Create new columns (average flight speed, whether late or not ...)**

#### **Average flight speed**

```python
#We are going to calculate the average flight speed and add it into our data set

airline_df["average_flight_speed(km/h)"] = ((airline_df["distance"] * 1.60934) / (airline_df["airtime"] / 60)).round(2)
airline_df.head()
```

#### **Flight was late or not**

```python
#We are going to construct a dichotomous variable. 1 if is late 0 on the contrary
#We add the new variable in our data set

def delayedFligt(x):
    if x > 0:
        return 1
    else:
        return 0

airline_df["delayed_flight"] = (airline_df["arrdelay"].apply(delayedFligt))
airline_df.head()
```

### **Table of airlines with the most accumulated arrears**

* We want to know the relative delayed flights. For that reason we distinguish the airline carriers per the total quantity of flights and his relative delays.

```python
# First, we want to know the number of flights per airline carrier and his %
airlines_flights = airline_df.uniquecarrier.value_counts()
airlines_percent = airline_df.uniquecarrier.value_counts(normalize=True).mul(100).round(2).astype(str) + '%'
carriers_df = pd.DataFrame({'annual flights': airlines_flights, 'percent of flights': airlines_percent})

# Second, we want to know the relative delayed flights per airline carrier

grouped = airline_df.groupby('uniquecarrier')
carrier_delays = (grouped['delayed_flight'].agg(np.sum))
carriers_df['annual delays'] = carrier_delays
carriers_df['percent of flights delayed'] = ((carriers_df['annual delays']/(carriers_df['annual flights']))*100).round(2)
carrier_delays = carriers_df.sort_values(by=['percent of flights delayed'],ascending=False).copy()
carrier_delays
```

**THE TOP 3 PERFORMERS FOR ARRIVAL DELAYS:**

    - US -> Puerto Rico International Airlines
    - WN -> Southwest Airlines
    
**THE WORST 3 PERFORMERS FOR ARRIVAL DELAYS:**

    - HA -> Hawaiian Airlines
    - YV -> Mesa Airlines
    - OH -> PSA Airlines

### **What are the longest flights? And the most backward?**

#### **Longest Flights**

```python
#DISTANCE (MILES)

grouped = airline_df.groupby(['uniquecarrier','flightnum','origin','dest']) #Select the interesting columns
longest_flights = (grouped['distance'].agg(np.mean)) #add 'distance' to the group and make the mean
longest_flights = longest_flights.sort_values(ascending=False) #reordering
longest_flights.head(10)
```

#### **Most delayed flights**

```python
# The bigger 'arrdelay' is, the bigger the delay

grouped = airline_df.groupby(['uniquecarrier','flightnum','origin','dest'])
delayed_flights = (grouped['arrdelay'].agg(np.mean))
delayed_flights = delayed_flights.sort_values(ascending=False)
delayed_flights.head(5)
```
